```python
import os
import pandas as pd
```

```python
img_dir = '/home/dongbeen/deeplearning/Competition/data/train_img/'
mask_dir = '/home/dongbeen/deeplearning/Competition/data/rotate_mask_image/'
```

```python
img_paths = [os.path.join(img_dir, img_name) for img_name in sorted(os.listdir(img_dir))]
mask_paths = [os.path.join(mask_dir, mask_name) for mask_name in sorted(os.listdir(mask_dir))]
```

```python
# 데이터프레임을 만듭니다.
df = pd.DataFrame({'img_path': img_paths, 'mask_path': mask_paths})
```

```python
# csv 파일로 저장합니다.
df.to_csv('image_mask_paths.csv', index=False)
```
